# 01 - 获取股票列表

**功能**: 从 Tushare 获取 A 股股票列表

```python
import os
import pandas as pd
import tushare as ts
import sqlite3
from pathlib import Path
from dotenv import load_dotenv

# 加载环境变量
PROJECT_ROOT = Path('../../').resolve()
load_dotenv(PROJECT_ROOT / '.env')
TUSHARE_TOKEN = os.getenv('TUSHARE_TOKEN')

# 初始化 Tushare
ts.set_token(TUSHARE_TOKEN)
pro = ts.pro_api()
print("✅ Tushare 初始化成功")

# 数据库连接
db_path = PROJECT_ROOT / 'data' / 'invest.db'
conn = sqlite3.connect(db_path)
print("✅ 数据库连接成功")
```

```python
# 获取 A 股股票列表
stock_list = pro.stock_basic(
    exchange='',
    list_status='L',
    fields='ts_code,symbol,name,area,industry,market,list_date'
)

print(f"✅ 获取到 {len(stock_list)} 只 A 股股票")
print("\n前 10 只股票：")
display(stock_list.head(10))
```

```python
# 保存到数据库
stock_list.to_sql('stock_list', conn, if_exists='replace', index=False)
print("✅ 股票列表已保存到数据库")

# 验证
count = pd.read_sql_query("SELECT COUNT(*) as count FROM stock_list", conn)['count'][0]
print(f"数据库中共有 {count} 只股票")
```
