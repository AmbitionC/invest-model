# Technical Factors - 技术因子计算主入口

**功能**: 统一执行所有技术因子计算

**因子列表**:
1. 布林带（Bollinger Bands）
2. MACD
3. RSI
4. 动量（Momentum）
5. 成交量（Volume）
6. 波动率（Volatility）
7. 偏离度（Deviation）

## 执行顺序

1. [01_calculate_bollinger.ipynb](./01_calculate_bollinger.ipynb) - 布林带因子
2. [02_calculate_macd_rsi.ipynb](./02_calculate_macd_rsi.ipynb) - MACD 和 RSI
3. [03_calculate_other_factors.ipynb](./03_calculate_other_factors.ipynb) - 其他因子
4. 04_save_factors.ipynb - 保存计算结果（待创建）

## 快速执行

```bash
jupyter nbconvert --to notebook --execute 01_calculate_bollinger.ipynb
jupyter nbconvert --to notebook --execute 02_calculate_macd_rsi.ipynb
jupyter nbconvert --to notebook --execute 03_calculate_other_factors.ipynb
```

## 核心设计

### 所有因子相对化（0-1 范围）

| 因子 | 相对化指标 | 含义 |
|------|-----------|------|
| 布林带 | `boll_bandwidth_pct` | 带宽分位数 |
| MACD | `macd_dif_pct` | DIF 分位数 |
| RSI | `rsi_pct` | RSI 分位数 |
| 动量 | `momentum_pct` | 动量分位数 |
| 成交量 | `volume_pct` | 成交量分位数 |
| 波动率 | `volatility_pct` | 波动率分位数 |
| 偏离度 | `deviation_pct` | 偏离度分位数 |

**好处**: 不同股票可以对比，不用绝对值判断
