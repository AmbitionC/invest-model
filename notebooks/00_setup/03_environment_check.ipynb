# 03 - 环境检查

**功能**: 验证所有配置是否正确

```python
import os
from pathlib import Path
import sqlite3
import pandas as pd

PROJECT_ROOT = Path('../../').resolve()

print("=" * 60)
print("环境检查清单")
print("=" * 60)

checks = []

# 1. 检查 .env 文件
env_path = PROJECT_ROOT / '.env'
if env_path.exists():
    checks.append((".env 文件", "✅"))
    TUSHARE_TOKEN = os.getenv('TUSHARE_TOKEN', '')
    if TUSHARE_TOKEN:
        checks.append(("Tushare Token", "✅"))
    else:
        checks.append(("Tushare Token", "❌"))
else:
    checks.append((".env 文件", "❌"))
    checks.append(("Tushare Token", "❌"))

# 2. 检查配置文件
config_path = PROJECT_ROOT / 'config' / 'config.yaml'
if config_path.exists():
    checks.append(("配置文件", "✅"))
else:
    checks.append(("配置文件", "❌"))

# 3. 检查数据库
db_path = PROJECT_ROOT / 'data' / 'invest.db'
if db_path.exists():
    checks.append(("数据库文件", "✅"))
    try:
        conn = sqlite3.connect(db_path)
        tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", conn)
        if len(tables) > 0:
            checks.append(("数据表", "✅"))
        else:
            checks.append(("数据表", "❌"))
        conn.close()
    except:
        checks.append(("数据表", "❌"))
else:
    checks.append(("数据库文件", "❌"))
    checks.append(("数据表", "❌"))

# 4. 检查依赖库
try:
    import tushare
    checks.append(("Tushare 库", "✅"))
except:
    checks.append(("Tushare 库", "❌"))

try:
    import plotly
    checks.append(("Plotly 库", "✅"))
except:
    checks.append(("Plotly 库", "❌"))

# 打印检查结果
for item, status in checks:
    print(f"{item:15} {status}")

print("=" * 60)

if all(status == "✅" for _, status in checks):
    print("🎉 所有检查通过！可以开始使用")
else:
    print("⚠️  有检查未通过，请修复后继续")
```
