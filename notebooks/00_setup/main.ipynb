# Setup - 环境配置主入口

**功能**: 统一执行所有 setup 步骤

**步骤**:
1. 环境配置
2. 数据库初始化
3. 环境检查

## 执行所有 Setup 步骤

按顺序执行以下 Notebook：
1. `01_environment_setup.ipynb` - 配置环境变量
2. `02_database_init.ipynb` - 初始化数据库
3. `03_environment_check.ipynb` - 验证配置

## 快速执行（推荐）

在终端运行：
```bash
jupyter nbconvert --to notebook --execute 01_environment_setup.ipynb
jupyter nbconvert --to notebook --execute 02_database_init.ipynb
jupyter nbconvert --to notebook --execute 03_environment_check.ipynb
```

## 或者手动执行

依次打开并执行：
- [01_environment_setup.ipynb](./01_environment_setup.ipynb)
- [02_database_init.ipynb](./02_database_init.ipynb)
